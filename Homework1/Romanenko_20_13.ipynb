```python
#Створіть дві матриці розміром 4x4, заповніть їх випадковими цілими числами в
#діапазоні від 1 до 5 та знайдіть різницю їх елементів.
import numpy as np

A = np.random.randint(1, 6, (4, 4))
B = np.random.randint(1, 6, (4, 4))

print("Матриця A:")
print(A)

print("\nМатриця B:")
print(B)

difference = A - B

print("\nРізниця елементів матриць (A - B):")
print(difference)
```
