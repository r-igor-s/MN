```python
#Створіть дві матриці 3x3, заповніть їх довільними цілими числами від 1  до 20 та
#знайдіть їх добуток.
import numpy as np

A = np.random.randint(1, 21, (3, 3))
B = np.random.randint(1, 21, (3, 3))

print("Матриця A:")
print(A)

print("\nМатриця B:")
print(B)

product = np.dot(A, B)

print("\nДобуток A * B:")
print(product)
```
