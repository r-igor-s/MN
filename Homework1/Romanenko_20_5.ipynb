```python
#Створіть два одновимірних масиви розміром 4, заповніть їх випадковими цілими
#числами від -10 до 10 та виконайте на них поелементні операції додавання, віднімання
#та множення.
import numpy as np

a = np.random.randint(-10, 11, 4)
b = np.random.randint(-10, 11, 4)

print("Масив A:", a)
print("Масив B:", b)

add_result = a + b
sub_result = a - b
mul_result = a * b

print("A + B =", add_result)
print("A - B =", sub_result)
print("A * B =", mul_result)
```
